# Replace commas in depth,porosity,density and permeability column with decimal point

```python
columns_to_replace =[ 'DEPTH_BASE', 'POROSITY_HORIZ_PCT', 'GRAIN_DENSITY_G_CM3', 'PERMEABILITY_HORIZ_1_PM' ]
df[columns_to_replace] = df[columns_to_replace].replace(',', '.', regex=True)
```

```python
df
```

# Replace all zero values in depth,porosity,density and permeability column with missing data because it is impossible to have zero value for such parameters

```python
columns_to_replace =[ 'DEPTH_BASE', 'POROSITY_HORIZ_PCT', 'GRAIN_DENSITY_G_CM3', 'PERMEABILITY_HORIZ_1_PM' ]
df[columns_to_replace] = df[columns_to_replace].replace('0', '', regex=True)
```

```python
df
```

# Replace dashes in all the column with missing data

```python
df = df.applymap(str)
df = df.apply(lambda x: x.str.replace('-', ''))
df = df.apply(pd.to_numeric, errors='ignore')
```

```python
df
```

# Replace missing data and blanks with value -99.25 which is used because it is a value that is outside of the expected range of valid measurements in many types of data.

```python
df.replace("", "-99.25", inplace=True)
```

```python
df
```

```python
df.replace("nan", "-99.25", inplace=True)
```

# Plot of porosity vs depth on each of the wells to better understand the geology of the subsurface

```python
import matplotlib.pyplot as plt


# Create a scatterplot of POROSITY_HORIZ_PCT and DEPTH_BASE based on the SITE_NAME column
SITE_NAME = df['SITE_NAME'].unique()
fig, axs = plt.subplots(len(SITE_NAME), figsize=(8, 6*len(SITE_NAME)))
for i, SITE_NAME in enumerate(SITE_NAME):
    SITE_NAME_df = df[df['SITE_NAME'] == SITE_NAME]
    axs[i].scatter(SITE_NAME_df['POROSITY_HORIZ_PCT'], SITE_NAME_df['DEPTH_BASE'])
    axs[i].set_title(SITE_NAME)
    axs[i].set_xlabel('POROSITY_HORIZ_PCT')
    axs[i].set_ylabel('DEPTH_BASE')
plt.tight_layout()


plt.show()
```
